# CMPE 256 Hierarchical Clustering(Agglomerative)

## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

## Reading and visualize the Data

```python
X = np.array ([[3,7],[2,6],[2,2],[5,2],[5,5],[5,8],[6,6],[8,4],[7,3],[10,6],[12,8]])
```

```python
plt.figure(figsize =(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1])

for x, y in zip(X[:, 0], X[:, 1]):
    plt.text(x, y, '({}, {})'.format(x, y))
plt.show()
```

## Single Link Cluster

```python
print ("Single Link Cluster")
linked = linkage(X, 'single')
plt.figure(figsize =(10, 7))
dendrogram(linked,
           orientation='top',
           labels = X,
           distance_sort = 'descending',
           show_leaf_counts = True)
plt.show()
```

## Complete Link Cluster

```python
print ("Complete Link Cluster")
linked_complete = linkage(X, 'complete')
plt.figure(figsize =(10, 7))
dendrogram(
    linked_complete ,
    labels = X,
    orientation='top',
    distance_sort  = 'descending',
    show_leaf_counts = True)
plt.show()
```

## Average Link Cluster

```python
print ("Average link cluster")
linked_average = linkage(X, 'average')
plt.figure(figsize =(10, 7))
dendrogram(
    linked_average,
    orientation='top',
    labels = X,
    distance_sort = 'descending',
    show_leaf_counts = True)
plt.show()
```

# CMPE 256 Starbucks Hot Beverages Cluster Assignment


### Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
```

## Reading the Data

```python
X = pd.read_csv("/content/drive/MyDrive/Dataset/StarbucksDataSet.csv") 
X
```

## Data Visualisation

```python
plt.figure(figsize =(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X.iloc[:,0],X.iloc[:,1])

for x, y in zip(X.iloc[:, 0], X.iloc[:, 1]):
    plt.text(x, y, '({}, {})'.format(x, y))
plt.show()
```

## Single Link Cluster

```python
print ("Single Link Cluster")
linked = linkage(X, 'single')
plt.figure(figsize =(10, 7))
dendrogram(linked,
           orientation='top',
           distance_sort = 'descending',
           show_leaf_counts = True)
plt.show()
```

```python
cluster_single = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='single')
cluster_single.fit_predict(X)
```

```python
plt.figure(figsize=(10, 7))
plt.xlabel("Popular Time Index")
plt.ylabel("Hot Beverages Index")
plt.scatter(X['PTIndex'], X['HBIndex'], c=cluster_single.labels_)
```

## Average Link Cluster

```python
print ("Average Link Cluster")
linked_average = linkage(X, 'average')
plt.figure(figsize =(10, 7))
dendrogram(linked_average,
           orientation='top',
           distance_sort = 'descending',
           show_leaf_counts = True)
plt.show()
```

```python
cluster_average = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='average')
cluster_average.fit_predict(X)
```

```python
plt.figure(figsize=(10, 7)) 
plt.xlabel("Popular Time Index")
plt.ylabel("Hot Beverages Index")
plt.scatter(X['PTIndex'], X['HBIndex'], c=cluster_average.labels_)
```

## Complete Link Cluster

```python
print ("Complete Link Cluster")
linked_complete = linkage(X, 'complete')
plt.figure(figsize =(10, 7))
dendrogram(linked_complete,
           orientation='top',
           distance_sort = 'descending',
           show_leaf_counts = True)
plt.show()
```

```python
cluster_complete = AgglomerativeClustering(n_clusters=3, linkage='complete')
cluster_complete.fit_predict(X)
```

```python
plt.figure(figsize=(10, 7)) 
plt.xlabel("Popular Time Index")
plt.ylabel("Hot Beverages Index")
plt.scatter(X['PTIndex'], X['HBIndex'], c=cluster_complete.labels_)
```
